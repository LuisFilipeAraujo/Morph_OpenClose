```python
img = cv2.imread("samoieda.jpg")
imgrgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(imgrgb)
```

```python
imggray = cv2.cvtColor(imgrgb, cv2.COLOR_BGR2GRAY)
cv2.imwrite("Cinza.jpg",imggray)

gray = cv2.imread("Cinza.jpg")
plt.imshow(gray)
```

```python
gray = cv2.cvtColor(imgrgb, cv2.COLOR_RGB2GRAY)
r = 128
imgBin = np.uint16(gray / r) *r
cv2.imwrite("bin.jpg",imgBin)
bin = cv2.imread("bin.jpg")

fig = plt.figure(figsize=(20,10))
ax1 = fig.add_subplot (131) 
plt.xticks([]), plt.yticks([])
plt.imshow(imgrgb)
plt.title("Recorte da imagem")


ax2 = fig.add_subplot (132) 
plt.xticks([]), plt.yticks([])
plt.imshow(gray, cmap= 'gray')
plt.title("CINZA")



ax3 = fig.add_subplot (133) 
plt.xticks([]), plt.yticks([])
plt.imshow(bin, cmap= 'gray')
plt.title("BINARIA")
```

## **ELEMENTO ESTRUTURANTE**

```python
#element_estr = cv2.getStructuringElement(cv2.CROSS,(21,21))
#element_estr = cv2.getStructuringElement(cv2.MORPH_RECT,(21,21))
element_estr = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(21,21))
```

```python
print(element_estr)
```

```python
imgProc1 = cv2.morphologyEx(bin, cv2.MORPH_OPEN,element_estr)
imgProc2 = cv2.morphologyEx(bin, cv2.MORPH_CLOSE,element_estr)
```

```python
fig =  plt.figure(figsize=(20,20))
ax1 = fig.add_subplot(141)
plt.imshow(imgrgb)
plt.title("Imagem Original")

ax2 = fig.add_subplot(142)
plt.imshow(bin)
plt.title("Imagem Binaria")

ax3 = fig.add_subplot(143)
plt.imshow(imgProc1)
plt.title("OPEN 21X21")

ax4 = fig.add_subplot(144)
plt.imshow(imgProc2)
plt.title("CLOSE 21X21")

plt.show()
```
